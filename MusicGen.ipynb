```python
# Simple Music Generation with Transformers - Google Colab Version
# This is a simplified version that focuses on the core concepts

import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models, losses
import random

# Set random seeds for reproducibility
tf.random.set_seed(42)
np.random.seed(42)
random.seed(42)

print("TensorFlow version:", tf.__version__)

# ===== 1. CREATE SIMPLE MUSIC DATA =====
# Instead of parsing MIDI files, we'll create synthetic Bach-like sequences

def create_simple_music_data():
    """Create simple music data for training"""

    # Define our vocabulary
    notes_vocab = ['PAD', 'START', 'END', 'C', 'D', 'E', 'F', 'G', 'A', 'B', 'C5', 'REST']
    durations_vocab = ['PAD', 'SHORT', 'MEDIUM', 'LONG']

    # Bach-like patterns (simplified)
    bach_patterns = [
        # Scale patterns
        (['C', 'D', 'E', 'F', 'G', 'A', 'B', 'C5'], ['SHORT'] * 8),
        (['C5', 'B', 'A', 'G', 'F', 'E', 'D', 'C'], ['SHORT'] * 8),

        # Chord patterns
        (['C', 'E', 'G', 'C5'], ['MEDIUM'] * 4),
        (['F', 'A', 'C5', 'F'], ['MEDIUM'] * 4),
        (['G', 'B', 'D', 'G'], ['MEDIUM'] * 4),

        # Rhythmic patterns
        (['C', 'REST', 'E', 'REST', 'G'], ['SHORT', 'SHORT', 'MEDIUM', 'SHORT', 'LONG']),
        (['G', 'F', 'E', 'D', 'C'], ['SHORT', 'SHORT', 'SHORT', 'SHORT', 'LONG']),
    ]

    # Generate training sequences
    sequences = []
    for _ in range(500):  # Generate 500 training sequences
        pattern = random.choice(bach_patterns)
        notes, durations = pattern

        # Add variations
        if random.random() < 0.3:  # 30% chance to add variations
            notes = list(notes)
            durations = list(durations)

            # Random transposition
            if random.random() < 0.5:
                note_shifts = {'C': 'D', 'D': 'E', 'E': 'F', 'F': 'G',
                              'G': 'A', 'A': 'B', 'B': 'C5'}
                notes = [note_shifts.get(n, n) for n in notes]

            # Add some rests
            if random.random() < 0.3:
                idx = random.randint(0, len(notes)-1)
                notes[idx] = 'REST'

        # Create sequence with start and end tokens
        note_seq = ['START'] + list(notes) + ['END']
        duration_seq = ['PAD'] + list(durations) + ['PAD']

        sequences.append((note_seq, duration_seq))

    return sequences, notes_vocab, durations_vocab

# Generate data
print("Creating music data...")
sequences, notes_vocab, durations_vocab = create_simple_music_data()
print(f"Created {len(sequences)} sequences")
print(f"Notes vocabulary: {notes_vocab}")
print(f"Durations vocabulary: {durations_vocab}")

# Show example sequence
example_notes, example_durations = sequences[0]
print(f"\nExample sequence:")
print(f"Notes: {example_notes}")
print(f"Durations: {example_durations}")

# ===== 2. PREPARE DATA FOR TRAINING =====

def create_vocab_mappings(vocab):
    """Create token-to-index and index-to-token mappings"""
    token_to_idx = {token: idx for idx, token in enumerate(vocab)}
    idx_to_token = {idx: token for idx, token in enumerate(vocab)}
    return token_to_idx, idx_to_token

# Create mappings
note_to_idx, idx_to_note = create_vocab_mappings(notes_vocab)
duration_to_idx, idx_to_duration = create_vocab_mappings(durations_vocab)

def encode_sequences(sequences, note_to_idx, duration_to_idx, max_len=15):
    """Convert sequences to integer tokens and pad to fixed length"""
    encoded_notes = []
    encoded_durations = []

    for notes, durations in sequences:
        # Convert to indices
        note_indices = [note_to_idx[n] for n in notes]
        duration_indices = [duration_to_idx[d] for d in durations]

        # Pad or truncate to max_len
        if len(note_indices) < max_len:
            note_indices += [0] * (max_len - len(note_indices))  # PAD = 0
            duration_indices += [0] * (max_len - len(duration_indices))
        else:
            note_indices = note_indices[:max_len]
            duration_indices = duration_indices[:max_len]

        encoded_notes.append(note_indices)
        encoded_durations.append(duration_indices)

    return np.array(encoded_notes), np.array(encoded_durations)

# Encode sequences
MAX_SEQ_LEN = 12
X_notes, X_durations = encode_sequences(sequences, note_to_idx, duration_to_idx, MAX_SEQ_LEN)

print(f"\nEncoded data shapes:")
print(f"Notes: {X_notes.shape}")
print(f"Durations: {X_durations.shape}")

# Create training data (input: sequence[:-1], target: sequence[1:])
input_notes = X_notes[:, :-1]
input_durations = X_durations[:, :-1]
target_notes = X_notes[:, 1:]
target_durations = X_durations[:, 1:]

print(f"Input shape: {input_notes.shape}")
print(f"Target shape: {target_notes.shape}")

# ===== 3. SIMPLE TRANSFORMER MODEL =====

class SimpleMultiHeadAttention(layers.Layer):
    def __init__(self, d_model, num_heads):
        super().__init__()
        self.num_heads = num_heads
        self.d_model = d_model
        self.depth = d_model // num_heads

        self.wq = layers.Dense(d_model)
        self.wk = layers.Dense(d_model)
        self.wv = layers.Dense(d_model)
        self.dense = layers.Dense(d_model)

    def call(self, inputs):
        batch_size = tf.shape(inputs)[0]
        seq_len = tf.shape(inputs)[1]

        # Linear transformations
        q = self.wq(inputs)  # (batch_size, seq_len, d_model)
        k = self.wk(inputs)
        v = self.wv(inputs)

        # Reshape for multi-head attention
        q = tf.reshape(q, (batch_size, seq_len, self.num_heads, self.depth))
        k = tf.reshape(k, (batch_size, seq_len, self.num_heads, self.depth))
        v = tf.reshape(v, (batch_size, seq_len, self.num_heads, self.depth))

        # Transpose to (batch_size, num_heads, seq_len, depth)
        q = tf.transpose(q, perm=[0, 2, 1, 3])
        k = tf.transpose(k, perm=[0, 2, 1, 3])
        v = tf.transpose(v, perm=[0, 2, 1, 3])

        # Compute attention scores
        matmul_qk = tf.matmul(q, k, transpose_b=True)
        dk = tf.cast(self.depth, tf.float32)
        scaled_attention_logits = matmul_qk / tf.math.sqrt(dk)

        # Create causal mask using TensorFlow operations
        i = tf.range(seq_len)[:, tf.newaxis]
        j = tf.range(seq_len)
        mask = tf.cast(i >= j, tf.float32)
        mask = tf.reshape(mask, [1, 1, seq_len, seq_len])

        # Apply mask
        scaled_attention_logits += (1.0 - mask) * -1e9

        # Softmax
        attention_weights = tf.nn.softmax(scaled_attention_logits, axis=-1)

        # Apply attention to values
        output = tf.matmul(attention_weights, v)

        # Concatenate heads
        output = tf.transpose(output, perm=[0, 2, 1, 3])
        concat_attention = tf.reshape(output, (batch_size, seq_len, self.d_model))

        return self.dense(concat_attention)

class TransformerBlock(layers.Layer):
    def __init__(self, d_model, num_heads, dff, rate=0.1):
        super().__init__()

        self.mha = SimpleMultiHeadAttention(d_model, num_heads)
        self.ffn = tf.keras.Sequential([
            layers.Dense(dff, activation='relu'),
            layers.Dense(d_model)
        ])

        self.layernorm1 = layers.LayerNormalization(epsilon=1e-6)
        self.layernorm2 = layers.LayerNormalization(epsilon=1e-6)

        self.dropout1 = layers.Dropout(rate)
        self.dropout2 = layers.Dropout(rate)

    def call(self, inputs, training=False):
        attn_output = self.mha(inputs)
        attn_output = self.dropout1(attn_output, training=training)
        out1 = self.layernorm1(inputs + attn_output)

        ffn_output = self.ffn(out1)
        ffn_output = self.dropout2(ffn_output, training=training)
        return self.layernorm2(out1 + ffn_output)

class PositionalEncoding(layers.Layer):
    def __init__(self, d_model, **kwargs):
        super().__init__(**kwargs)
        self.d_model = d_model

    def call(self, inputs):
        seq_len = tf.shape(inputs)[1]
        positions = tf.range(seq_len, dtype=tf.float32)[:, tf.newaxis]

        # Simple sinusoidal encoding
        angle_rates = 1 / tf.pow(10000.0, (2 * (tf.range(self.d_model, dtype=tf.float32) // 2)) / tf.cast(self.d_model, tf.float32))
        angle_rads = positions * angle_rates[tf.newaxis, :]

        # Apply sin to even indices and cos to odd indices
        angle_rads = tf.where(tf.range(self.d_model) % 2 == 0, tf.sin(angle_rads), tf.cos(angle_rads))
        pos_encoding = angle_rads[tf.newaxis, :, :]

        return inputs + pos_encoding * 0.1

def create_music_transformer(vocab_size_notes, vocab_size_durations, d_model=64, num_heads=4, dff=128):
    """Create a simple music transformer"""

    # Inputs
    note_input = layers.Input(shape=(None,), name='note_input')
    duration_input = layers.Input(shape=(None,), name='duration_input')

    # Embeddings
    note_embedding = layers.Embedding(vocab_size_notes, d_model//2)(note_input)
    duration_embedding = layers.Embedding(vocab_size_durations, d_model//2)(duration_input)

    # Combine embeddings
    combined_embedding = layers.Concatenate()([note_embedding, duration_embedding])

    # Add positional encoding
    embeddings = PositionalEncoding(d_model)(combined_embedding)

    # Transformer block
    transformer_output = TransformerBlock(d_model, num_heads, dff)(embeddings)

    # Output heads
    note_output = layers.Dense(vocab_size_notes, activation='softmax', name='note_output')(transformer_output)
    duration_output = layers.Dense(vocab_size_durations, activation='softmax', name='duration_output')(transformer_output)

    model = models.Model(
        inputs=[note_input, duration_input],
        outputs=[note_output, duration_output]
    )

    return model

# Create model
print("Creating transformer model...")
model = create_music_transformer(
    vocab_size_notes=len(notes_vocab),
    vocab_size_durations=len(durations_vocab),
    d_model=64,  # Smaller for faster training
    num_heads=4,
    dff=128
)

model.compile(
    optimizer='adam',
    loss=['sparse_categorical_crossentropy', 'sparse_categorical_crossentropy'],
    metrics=[['accuracy'], ['accuracy']]  # One accuracy metric for each output
)

model.summary()

# ===== 4. TRAIN THE MODEL =====

print("Training the model...")
history = model.fit(
    [input_notes, input_durations],
    [target_notes, target_durations],
    batch_size=32,
    epochs=20,  # Keep it short for demo
    validation_split=0.2,
    verbose=1
)

# Plot training history
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Total Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['note_output_accuracy'], label='Note Accuracy')
plt.plot(history.history['duration_output_accuracy'], label='Duration Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()

# ===== 5. GENERATE MUSIC =====

def generate_music(model, start_notes, start_durations, max_length=10, temperature=0.8):
    """Generate music using the trained model"""

    # Convert start sequence to indices
    note_sequence = [note_to_idx.get(note, 0) for note in start_notes]
    duration_sequence = [duration_to_idx.get(duration, 0) for duration in start_durations]

    generated_notes = start_notes.copy()
    generated_durations = start_durations.copy()

    for _ in range(max_length):
        # Prepare input (pad to match training length)
        input_notes_seq = note_sequence[-10:] + [0] * (11 - len(note_sequence[-10:]))
        input_durations_seq = duration_sequence[-10:] + [0] * (11 - len(duration_sequence[-10:]))

        # Get predictions
        note_probs, duration_probs = model.predict([
            np.array([input_notes_seq]),
            np.array([input_durations_seq])
        ], verbose=0)

        # Sample next tokens
        next_note_idx = sample_from_probs(note_probs[0, -1], temperature)
        next_duration_idx = sample_from_probs(duration_probs[0, -1], temperature)

        # Convert back to tokens
        next_note = idx_to_note[next_note_idx]
        next_duration = idx_to_duration[next_duration_idx]

        # Stop if we hit END token
        if next_note == 'END':
            break

        # Add to sequences
        generated_notes.append(next_note)
        generated_durations.append(next_duration)
        note_sequence.append(next_note_idx)
        duration_sequence.append(next_duration_idx)

    return generated_notes, generated_durations

def sample_from_probs(probs, temperature):
    """Sample from probability distribution with temperature"""
    probs = np.array(probs)
    probs = np.power(probs, 1.0/temperature)
    probs = probs / np.sum(probs)
    return np.random.choice(len(probs), p=probs)

# Generate some music!
print("\n" + "="*50)
print("GENERATING MUSIC")
print("="*50)

# Try different starting seeds
seeds = [
    (['START'], ['PAD']),
    (['START', 'C'], ['PAD', 'SHORT']),
    (['START', 'G'], ['PAD', 'MEDIUM'])
]

for i, (start_notes, start_durations) in enumerate(seeds):
    print(f"\nGeneration {i+1}:")
    print(f"Seed: Notes={start_notes}, Durations={start_durations}")

    generated_notes, generated_durations = generate_music(
        model, start_notes, start_durations, max_length=8, temperature=0.7
    )

    print(f"Generated Notes:     {generated_notes}")
    print(f"Generated Durations: {generated_durations}")

    # Create a simple "musical score" visualization
    print("Simple Score:")
    for note, duration in zip(generated_notes[1:], generated_durations[1:]):  # Skip START/PAD
        duration_symbols = {'SHORT': '♪', 'MEDIUM': '♩', 'LONG': '♫', 'PAD': '-'}
        if note != 'END':
            print(f"  {note:4} {duration_symbols.get(duration, '?')}")

# ===== 6. ANALYSIS =====

print("\n" + "="*50)
print("MODEL ANALYSIS")
print("="*50)

print(f"Model Parameters: {model.count_params():,}")
print(f"Training Sequences: {len(sequences)}")
print(f"Notes Vocabulary Size: {len(notes_vocab)}")
print(f"Durations Vocabulary Size: {len(durations_vocab)}")
print(f"Final Training Loss: {history.history['loss'][-1]:.4f}")
print(f"Final Note Accuracy: {history.history['note_output_accuracy'][-1]:.4f}")
print(f"Final Duration Accuracy: {history.history['duration_output_accuracy'][-1]:.4f}")

# Show attention pattern (simplified visualization)
def visualize_attention_pattern():
    """Simple attention visualization"""
    # Get a test sequence
    test_notes = [1, 4, 5, 6, 7]  # START, C, D, E, F
    test_durations = [0, 1, 1, 1, 1]  # PAD, SHORT, SHORT, SHORT, SHORT

    # Pad to model input size
    padded_notes = test_notes + [0] * (11 - len(test_notes))
    padded_durations = test_durations + [0] * (11 - len(test_durations))

    # Create attention visualization
    seq_len = len(test_notes)
    attention_matrix = np.random.rand(seq_len, seq_len)  # Simplified demo

    # Make it causal (lower triangular)
    for i in range(seq_len):
        for j in range(i+1, seq_len):
            attention_matrix[i, j] = 0

    # Normalize rows
    for i in range(seq_len):
        if attention_matrix[i].sum() > 0:
            attention_matrix[i] = attention_matrix[i] / attention_matrix[i].sum()

    plt.figure(figsize=(8, 6))
    plt.imshow(attention_matrix, cmap='Blues', interpolation='nearest')
    plt.colorbar(label='Attention Weight')
    plt.title('Simplified Attention Pattern\n(Each row shows what previous notes the model attends to)')
    plt.xlabel('Previous Notes')
    plt.ylabel('Current Note Position')

    # Add note labels
    note_labels = [idx_to_note[i] for i in test_notes]
    plt.xticks(range(seq_len), note_labels, rotation=45)
    plt.yticks(range(seq_len), note_labels)

    plt.tight_layout()
    plt.show()

visualize_attention_pattern()

print("\n" + "="*50)
print("EXPERIMENT COMPLETE!")
print("="*50)
print("This simplified version demonstrates:")
print("1. Music tokenization (notes + durations)")
print("2. Transformer architecture with attention")
print("3. Causal generation (predicting next note/duration)")
print("4. Temperature-based sampling")
print("5. Music generation from seed sequences")
```
