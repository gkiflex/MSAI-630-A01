```python
# RealNVP Neural Network Implementation
# This notebook implements a RealNVP (Real-valued Non-Volume Preserving) transformation
# to learn complex data distributions and transform them into simple Gaussian distributions

# Install required packages (run this cell first in Google Colab)
!pip install tensorflow tensorflow-probability matplotlib scikit-learn numpy

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
import tensorflow as tf
from tensorflow.keras import layers, models, regularizers, metrics, optimizers, callbacks
import tensorflow_probability as tfp

print(f"TensorFlow version: {tf.__version__}")
print(f"TensorFlow Probability version: {tfp.__version__}")

# =============================================================================
# 0. PARAMETERS
# =============================================================================

COUPLING_DIM = 256      # Dimension of coupling layers
COUPLING_LAYERS = 2     # Number of coupling layers
INPUT_DIM = 2          # Input dimension (2D for moons dataset)
REGULARIZATION = 0.01  # L2 regularization strength
BATCH_SIZE = 256       # Batch size for training
EPOCHS = 600          # Number of training epochs (increased from 300 to 600)

print(f"Training Configuration:")
print(f"- Coupling Dimension: {COUPLING_DIM}")
print(f"- Coupling Layers: {COUPLING_LAYERS}")
print(f"- Input Dimension: {INPUT_DIM}")
print(f"- Regularization: {REGULARIZATION}")
print(f"- Batch Size: {BATCH_SIZE}")
print(f"- Epochs: {EPOCHS}")

# =============================================================================
# 1. DATA PREPARATION
# =============================================================================

# Generate the moons dataset
print("\nGenerating moons dataset...")
data = datasets.make_moons(30000, noise=0.05)[0].astype("float32")

# Normalize the data
norm = layers.Normalization()
norm.adapt(data)
normalized_data = norm(data)

# Visualize the original dataset
plt.figure(figsize=(8, 6))
plt.scatter(normalized_data.numpy()[:, 0], normalized_data.numpy()[:, 1],
           c="blue", alpha=0.6, s=1)
plt.title("Original Moons Dataset (Normalized)", fontsize=14)
plt.xlabel("x₁")
plt.ylabel("x₂")
plt.grid(True, alpha=0.3)
plt.show()

print(f"Dataset shape: {normalized_data.shape}")
print(f"Data range: x₁ ∈ [{normalized_data[:, 0].numpy().min():.2f}, {normalized_data[:, 0].numpy().max():.2f}], "
      f"x₂ ∈ [{normalized_data[:, 1].numpy().min():.2f}, {normalized_data[:, 1].numpy().max():.2f}]")

# =============================================================================
# 2. CUSTOM COUPLING LAYER
# =============================================================================

def Coupling(input_dim, coupling_dim, reg):
    """
    Custom Coupling layer for RealNVP.

    Creates two separate networks:
    - Scale network (s): outputs scaling factors with tanh activation
    - Translation network (t): outputs translation values with linear activation

    Args:
        input_dim: Dimension of input data
        coupling_dim: Hidden dimension for coupling networks
        reg: L2 regularization strength

    Returns:
        Keras Model with two outputs [scale, translation]
    """
    input_layer = layers.Input(shape=(input_dim,))

    # Scale network (s) - outputs scaling factors
    s_layer_1 = layers.Dense(
        coupling_dim, activation="relu",
        kernel_regularizer=regularizers.l2(reg)
    )(input_layer)
    s_layer_2 = layers.Dense(
        coupling_dim, activation="relu",
        kernel_regularizer=regularizers.l2(reg)
    )(s_layer_1)
    s_layer_3 = layers.Dense(
        coupling_dim, activation="relu",
        kernel_regularizer=regularizers.l2(reg)
    )(s_layer_2)
    s_layer_4 = layers.Dense(
        coupling_dim, activation="relu",
        kernel_regularizer=regularizers.l2(reg)
    )(s_layer_3)
    s_layer_5 = layers.Dense(
        input_dim, activation="tanh",  # tanh for bounded scaling
        kernel_regularizer=regularizers.l2(reg)
    )(s_layer_4)

    # Translation network (t) - outputs translation values
    t_layer_1 = layers.Dense(
        coupling_dim, activation="relu",
        kernel_regularizer=regularizers.l2(reg)
    )(input_layer)
    t_layer_2 = layers.Dense(
        coupling_dim, activation="relu",
        kernel_regularizer=regularizers.l2(reg)
    )(t_layer_1)
    t_layer_3 = layers.Dense(
        coupling_dim, activation="relu",
        kernel_regularizer=regularizers.l2(reg)
    )(t_layer_2)
    t_layer_4 = layers.Dense(
        coupling_dim, activation="relu",
        kernel_regularizer=regularizers.l2(reg)
    )(t_layer_3)
    t_layer_5 = layers.Dense(
        input_dim, activation="linear",  # linear for unbounded translation
        kernel_regularizer=regularizers.l2(reg)
    )(t_layer_4)

    return models.Model(inputs=input_layer, outputs=[s_layer_5, t_layer_5])

print("✓ Custom Coupling layer defined")

# =============================================================================
# 3. REALNVP MODEL
# =============================================================================

class RealNVP(models.Model):
    """
    Real-valued Non-Volume Preserving (RealNVP) transformation model.

    This model learns to transform complex data distributions into simple
    Gaussian distributions through a series of invertible coupling transformations.
    """

    def __init__(self, input_dim, coupling_layers, coupling_dim, regularization):
        super(RealNVP, self).__init__()
        self.coupling_layers = coupling_layers

        # Target distribution (standard multivariate normal)
        self.distribution = tfp.distributions.MultivariateNormalDiag(
            loc=[0.0, 0.0], scale_diag=[1.0, 1.0]
        )

        # Alternating masks for coupling layers
        # Each mask determines which dimensions are transformed
        self.masks = np.array(
            [[0, 1], [1, 0]] * (coupling_layers // 2), dtype="float32"
        )

        # Loss tracking
        self.loss_tracker = metrics.Mean(name="loss")

        # Create coupling layer networks
        self.layers_list = [
            Coupling(input_dim, coupling_dim, regularization)
            for i in range(coupling_layers)
        ]

        print(f"RealNVP Model initialized with {coupling_layers} coupling layers")
        print(f"Masks shape: {self.masks.shape}")

    @property
    def metrics(self):
        return [self.loss_tracker]

    def call(self, x, training=True):
        """
        Forward/inverse pass through the RealNVP network.

        Args:
            x: Input tensor
            training: If True, performs forward pass (data→latent)
                     If False, performs inverse pass (latent→data)

        Returns:
            Transformed tensor and log determinant of Jacobian
        """
        log_det_inv = 0
        direction = 1 if training else -1

        # Apply coupling transformations
        layer_range = range(self.coupling_layers)
        if not training:
            layer_range = reversed(layer_range)

        for i in layer_range:
            # Apply mask to select which dimensions to keep unchanged
            x_masked = x * self.masks[i]
            reversed_mask = 1 - self.masks[i]

            # Get scale and translation from coupling networks
            s, t = self.layers_list[i](x_masked)
            s *= reversed_mask  # Apply mask to scale
            t *= reversed_mask  # Apply mask to translation

            # Coupling transformation
            gate = (direction - 1) / 2
            x = (
                reversed_mask * (x * tf.exp(direction * s) + direction * t * tf.exp(gate * s))
                + x_masked
            )

            # Accumulate log determinant for probability computation
            log_det_inv += gate * tf.reduce_sum(s, axis=1)

        return x, log_det_inv

    def log_loss(self, x):
        """
        Compute negative log-likelihood loss.

        The loss is -log p(x) where p(x) is computed using the change of variables formula.
        """
        # Transform data to latent space
        y, logdet = self(x)

        # Compute log probability in latent space plus log determinant
        log_likelihood = self.distribution.log_prob(y) + logdet

        # Return negative log-likelihood (to minimize)
        return -tf.reduce_mean(log_likelihood)

    def train_step(self, data):
        """Custom training step."""
        with tf.GradientTape() as tape:
            loss = self.log_loss(data)

        # Compute gradients and update weights
        g = tape.gradient(loss, self.trainable_variables)
        self.optimizer.apply_gradients(zip(g, self.trainable_variables))

        # Update loss tracker
        self.loss_tracker.update_state(loss)
        return {"loss": self.loss_tracker.result()}

    def test_step(self, data):
        """Custom test step."""
        loss = self.log_loss(data)
        self.loss_tracker.update_state(loss)
        return {"loss": self.loss_tracker.result()}

# Create the RealNVP model
model = RealNVP(
    input_dim=INPUT_DIM,
    coupling_layers=COUPLING_LAYERS,
    coupling_dim=COUPLING_DIM,
    regularization=REGULARIZATION,
)

print("✓ RealNVP model created successfully")

# =============================================================================
# 4. TRAINING SETUP AND CALLBACKS
# =============================================================================

# Compile the model
model.compile(optimizer=optimizers.Adam(learning_rate=0.0001))
print("✓ Model compiled with Adam optimizer")

# Create visualization callback
class ImageGenerator(callbacks.Callback):
    """Callback to generate and display samples during training."""

    def __init__(self, num_samples):
        self.num_samples = num_samples
        self.epoch_losses = []

    def generate(self):
        """Generate samples and transformations."""
        # Transform data to latent space (forward pass)
        z, _ = model(normalized_data)

        # Sample from latent distribution and transform to data space (inverse pass)
        samples = model.distribution.sample(self.num_samples)
        x, _ = model.predict(samples, verbose=0)

        return x, z, samples

    def display(self, x, z, samples, epoch=None):
        """Display the four key plots."""
        f, axes = plt.subplots(2, 2, figsize=(12, 10))

        # Original data space
        axes[0, 0].scatter(normalized_data[:, 0], normalized_data[:, 1],
                          color="red", s=1, alpha=0.6)
        axes[0, 0].set_title("Original Data Space (X)", fontsize=12, fontweight='bold')
        axes[0, 0].set_xlabel("x₁")
        axes[0, 0].set_ylabel("x₂")
        axes[0, 0].set_xlim([-3, 3])
        axes[0, 0].set_ylim([-3, 3])
        axes[0, 0].grid(True, alpha=0.3)

        # Data transformed to latent space
        axes[0, 1].scatter(z[:, 0], z[:, 1], color="red", s=1, alpha=0.6)
        axes[0, 1].set_title("Data → Latent Space f(X)", fontsize=12, fontweight='bold')
        axes[0, 1].set_xlabel("z₁")
        axes[0, 1].set_ylabel("z₂")
        axes[0, 1].set_xlim([-3, 3])
        axes[0, 1].set_ylim([-3, 3])
        axes[0, 1].grid(True, alpha=0.3)

        # Samples from latent distribution
        axes[1, 0].scatter(samples[:, 0], samples[:, 1], color="green", s=1, alpha=0.6)
        axes[1, 0].set_title("Latent Space Samples (Z)", fontsize=12, fontweight='bold')
        axes[1, 0].set_xlabel("z₁")
        axes[1, 0].set_ylabel("z₂")
        axes[1, 0].set_xlim([-3, 3])
        axes[1, 0].set_ylim([-3, 3])
        axes[1, 0].grid(True, alpha=0.3)

        # Generated samples in data space
        axes[1, 1].scatter(x[:, 0], x[:, 1], color="green", s=1, alpha=0.6)
        axes[1, 1].set_title("Generated Samples g(Z)", fontsize=12, fontweight='bold')
        axes[1, 1].set_xlabel("x₁")
        axes[1, 1].set_ylabel("x₂")
        axes[1, 1].set_xlim([-3, 3])
        axes[1, 1].set_ylim([-3, 3])
        axes[1, 1].grid(True, alpha=0.3)

        plt.suptitle(f"RealNVP Transformations" + (f" - Epoch {epoch}" if epoch else ""),
                    fontsize=14, fontweight='bold')
        plt.tight_layout()
        plt.show()

    def on_epoch_end(self, epoch, logs=None):
        """Display results every 50 epochs and track loss."""
        if logs:
            self.epoch_losses.append(logs['loss'])

        if epoch % 50 == 0:
            print(f"\n--- Epoch {epoch} Results ---")
            x, z, samples = self.generate()
            self.display(x, z, samples, epoch)

# Create callbacks
img_generator_callback = ImageGenerator(num_samples=5000)

print("✓ Training callbacks created")

# =============================================================================
# 5. TRAIN THE MODEL
# =============================================================================

print(f"\n{'='*60}")
print(f"STARTING TRAINING FOR {EPOCHS} EPOCHS")
print(f"{'='*60}")

# Train the model
history = model.fit(
    normalized_data,
    batch_size=BATCH_SIZE,
    epochs=EPOCHS,
    callbacks=[img_generator_callback],
    verbose=1
)

print(f"\n{'='*60}")
print("TRAINING COMPLETED!")
print(f"{'='*60}")

# =============================================================================
# 6. RESULTS ANALYSIS
# =============================================================================

# Plot training loss curve
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], 'b-', linewidth=2, label='Training Loss')
plt.title('Training Loss Curve', fontsize=14, fontweight='bold')
plt.xlabel('Epoch')
plt.ylabel('Negative Log-Likelihood')
plt.grid(True, alpha=0.3)
plt.legend()

plt.subplot(1, 2, 2)
# Plot last 100 epochs for better detail
start_epoch = max(0, len(history.history['loss']) - 100)
plt.plot(range(start_epoch, len(history.history['loss'])),
         history.history['loss'][start_epoch:], 'r-', linewidth=2)
plt.title('Training Loss (Last 100 Epochs)', fontsize=14, fontweight='bold')
plt.xlabel('Epoch')
plt.ylabel('Negative Log-Likelihood')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Display final results
print("\n" + "="*60)
print("FINAL RESULTS")
print("="*60)

x, z, samples = img_generator_callback.generate()
img_generator_callback.display(x, z, samples)

# Print training statistics
final_loss = history.history['loss'][-1]
min_loss = min(history.history['loss'])
min_loss_epoch = np.argmin(history.history['loss'])

print(f"\nTraining Statistics:")
print(f"- Final Loss: {final_loss:.4f}")
print(f"- Minimum Loss: {min_loss:.4f} (at epoch {min_loss_epoch})")
print(f"- Total Parameters: {model.count_params():,}")

# Analyze the quality of the transformation
print(f"\nTransformation Quality Analysis:")
print(f"- Original data shape: {normalized_data.shape}")
print(f"- Latent space statistics:")
print(f"  - Mean: [{z[:, 0].numpy().mean():.3f}, {z[:, 1].numpy().mean():.3f}]")
print(f"  - Std:  [{z[:, 0].numpy().std():.3f}, {z[:, 1].numpy().std():.3f}]")
print(f"- Generated samples statistics:")
print(f"  - Mean: [{x[:, 0].mean():.3f}, {x[:, 1].mean():.3f}]")
print(f"  - Std:  [{x[:, 0].std():.3f}, {x[:, 1].std():.3f}]")

print("\n" + "="*60)
print("ANALYSIS COMPLETE!")
print("="*60)
```
